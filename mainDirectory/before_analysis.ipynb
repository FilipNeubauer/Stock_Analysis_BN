```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# loading data
df = pd.read_csv("words_for_analysis.csv")
```

```python
# count of words that occur n times
def number_of_words(n):
    for i in range(1, n+1):
       print("Number of words that occur " + str(i) + ":", len(df[df['count'] == i])) 

number_of_words(10)
```

```python
# histogram of n most occurences
n = 200

plt.figure(figsize=(12, 6))
plt.hist(x=df[df["count"] <n]["count"], edgecolor='black', bins=n) 
plt.xlabel('Frequency of occurrence')
plt.ylabel('Count of words')
plt.title('Occurence of words')

# Display the histogram
plt.show()
```

```python
# histogram of n most occurences with y log
n = len(df)

plt.figure(figsize=(12, 6))
plt.hist(x=df[df["count"] <n]["count"], edgecolor='black', bins=n)
plt.yscale('log')
plt.xlabel('Frequency of occurrence')
plt.ylabel('Count of words log')
plt.title('Occurence of words')

# Display the histogram
plt.show()
```

```python
# words wiht highest occurence
n = 10
n_largest = df.nlargest(n, "count")
print(n_largest)
```
